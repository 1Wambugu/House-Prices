```python
# Task 3: Explore Correlations

## Correlation with SalePrice
correlation_matrix = train_data.corr()
correlation_saleprice = correlation_matrix['SalePrice'].sort_values(ascending=False)
# Display top positive and negative correlations
print("\nTop positive correlations with SalePrice:")
print(correlation_saleprice.head(5))
print("\nTop negative correlations with SalePrice:")
print(correlation_saleprice.tail(5))
```

```python
# Calculate correlations with SalePrice
correlation_matrix = train_data.corr()
correlation_saleprice = correlation_matrix['SalePrice'].sort_values(ascending=False)

# Extract the top features with the strongest positive correlations
strongest_positive_corr = correlation_saleprice.head(3)  # Change 3 to the number of features you want to display

# Plot strongest positive correlations
plt.figure(figsize=(14, 10))
for i, feature in enumerate(strongest_positive_corr.index):
    plt.subplot(2, 3, i + 1)
    sns.scatterplot(x=feature, y='SalePrice', data=train_data, color='blue')
    plt.title(f'Relationship between {feature} and SalePrice')

plt.tight_layout()
plt.show()
```

```python
# Extract the top features with the strongest negative correlations
strongest_negative_corr = correlation_saleprice.tail(3)  # Change 3 to the number of features you want to display

# Plot strongest negative correlations
plt.figure(figsize=(14, 10))
for i, feature in enumerate(strongest_negative_corr.index):
    plt.subplot(2, 3, i + 1)
    sns.scatterplot(x=feature, y='SalePrice', data=train_data, color='red')  # You can choose a color of your preference
    plt.title(f'Relationship between {feature} and SalePrice')

plt.tight_layout()
plt.show()
```

```python
# Heatmap for correlations
plt.figure(figsize=(20, 10))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Task 4: Engineer and Explore a New Feature

## Age Feature
train_data['Age'] = train_data['YrSold'] - train_data['YearBuilt']

# Plot relationship between 'Age' and 'SalePrice'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='SalePrice', data=train_data)
plt.title('Relationship between Age and SalePrice')
plt.xlabel('Age')
plt.ylabel('SalePrice')
plt.show()
```
