```python
#part 1 Explore Data Distribution
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 1: Data Understanding
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
# Summary statistics for the training dataset
print(train_data.describe())
```

```python
test_data = pd.read_csv('test.csv')
print(test_data.describe())
```

```python
# Summary Statistics
summary_stats = train_data[['SalePrice', 'TotRmsAbvGrd', 'OverallCond']].describe()
plt.figure(figsize=(15, 5))
plt.subplot(131)
sns.histplot(train_data['SalePrice'], kde=True, color='blue')
plt.xlabel('Sale Price')
plt.title('Sale Price Distribution')
```

```python
# TotRmsAbvGrd Distribution
plt.subplot(132)
sns.histplot(train_data['TotRmsAbvGrd'], kde=True, color='blue')
plt.xlabel('Total Rooms Above Grade')
plt.title('Total Rooms Above Grade Distribution')
```

```python
# OverallCond Distribution
plt.subplot(133)
sns.histplot(train_data['OverallCond'], kde=True, color='orange')
plt.xlabel('Overall Condition')
plt.title('Overall Condition Distribution')

plt.tight_layout()
plt.show()
```

```python
# Interpretive Text
print("Summary Statistics:")
print(summary_stats)
print("\nInterpretation:")
print("Sale Price:")
print("The 'SalePrice' distribution shows a right-skewed distribution, indicating that most houses have lower prices, but some houses are priced significantly higher.")

print("\nTotal Rooms Above Grade:")
print("The 'TotRmsAbvGrd' distribution is approximately normal, with the majority of houses having around 6-7 rooms above grade.")

print("\nOverall Condition:")
print("The 'OverallCond' distribution indicates that most houses have an overall condition rating close to 5, which is considered average.")
```

```python
#Part2 Explore Differences between Subsets
# List of unique 'OverallCond' values
unique_conditions = train_data['OverallCond'].unique()

# Create subplots for each 'OverallCond' value
plt.figure(figsize=(15, 10))

for i, cond in enumerate(unique_conditions):
    plt.subplot(3, 3, i + 1)
    subset = train_data[train_data['OverallCond'] == cond]
    sns.histplot(subset['SalePrice'], kde=True, bins=20, color='blue')
    plt.xlabel('Sale Price')
    plt.title(f'OverallCond {cond}')

plt.tight_layout()
plt.show()
```

```python
for cond in unique_conditions:
    subset = train_data[train_data['OverallCond'] == cond]
    unique_prices = subset['SalePrice'].unique()
    print(f'OverallCond {cond} has {len(unique_prices)} unique SalePrice values.')
```

```python
#part3
#Explore Correlations
# Calculate the correlation matrix
correlation_matrix = train_data.corr()

# Find the features with the highest positive and negative correlations with 'SalePrice'
strongest_positive_corr = correlation_matrix['SalePrice'].sort_values(ascending=False).head(5)
strongest_negative_corr = correlation_matrix['SalePrice'].sort_values(ascending=True).head(5)

# Print the top correlated features
print("Features with the Strongest Positive Correlations:")
print(strongest_positive_corr)
```

```python
# Strongest Positive Correlations
plt.figure(figsize=(14, 10))
for i, feature in enumerate(strongest_positive_corr.index):
    plt.subplot(2, 3, i + 1)
    sns.scatterplot(x=feature, y='SalePrice', data=train_data, color='blue')
    plt.title(f'{feature} vs. SalePrice')
```

```python
#Part 4
# Engineer and Explore a New Feature 
# Calculate the age feature
train_data['Age'] = train_data['YrSold'] - train_data['YearBuilt']

# Create a scatter plot to visualize the relationship between 'Age' and 'SalePrice'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='SalePrice', data=train_data, color='blue')
plt.xlabel('Age')
plt.ylabel('Sale Price')
plt.title('Age vs. Sale Price')
plt.show()
```

```python
#polished visualizations of relevant findings
#Distribution of Sale Price:
plt.figure(figsize=(10, 6))
sns.histplot(train_data['SalePrice'], kde=True, color='blue')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.title('Distribution of Sale Price')
plt.show()
```

```python
#Relationship between Overall Condition and Sale Price:
plt.figure(figsize=(10, 6))
sns.boxplot(x='OverallCond', y='SalePrice', data=train_data, palette='viridis')
plt.xlabel('Overall Condition')
plt.ylabel('Sale Price')
plt.title('Relationship between Overall Condition and Sale Price')
plt.show()
```
